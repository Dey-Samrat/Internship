```python
# Q.4 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: 
# Brand 
# Product Description 
# Price 

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("http://www.flipkart.com/")
product=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
product.send_keys('sunglasses')
search=driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()

brand=[]
description=[]
price=[]
discount=[]

start=0
end=3

for page in range(start,end):
    brand_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand_tags:
        name=i.text
        brand.append(name)
        
    description_tags=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in description_tags:
        desc=i.text
        description.append(desc)
    
    price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in price_tags:
        cost=i.text
        price.append(cost)
        
    discount_tags=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]/span')
    for i in discount_tags:
        disc=i.text
        discount.append(disc)
    
    next_button=driver.find_element(By.CLASS_NAME,"_1LKTO3")
    next_button.click()
    time.sleep(3)
    
df=pd.DataFrame({'Brand':brand,'Description':description,'Price':price,'Discount':discount})
df.head(100)
```

```python
# Q.5 Scrape 100 reviews data from flipkart.com for iphone11 phone.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")

rating=[]
summary=[]
full=[]

start=0
end=10

for page in range(start,end):
    rating_tags=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tags:
        rat=i.text
        rating.append(rat)
        
    summary_tags=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in summary_tags:
        rev=i.text
        summary.append(rev)
    
    full_tags=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in full_tags:
        ful=i.text
        full.append(ful)
    
    next_button=driver.find_element(By.CLASS_NAME,"_1LKTO3")
    next_button.click()
    time.sleep(3)
    
df=pd.DataFrame({'Rating':rating,'Review Summary':summary,'Full Review':full})
df.head(100)
```

```python
# Q.8 A python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps: 
# 1. First get the webpagehttps://www.azquotes.com/ 
# 2. Click on Top Quotes
# 3. Than scrap a) Quote b) Author c) Type Of Quotes

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")
top=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top.click()

quotes=[]
authors=[]
types=[]

start=0
end=9

for page in range(start,end):
    quotes_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quotes_tags:
        quote=i.text
        quotes.append(quote)
    authors_tags=driver.find_elements(By.CLASS_NAME,"author")
    for i in authors_tags:
        author=i.text
        authors.append(author)
    types_tags=driver.find_elements(By.CLASS_NAME,"tags")
    for i in types_tags:
        typ=i.text
        types.append(typ)
    
    next_button=driver.find_element(By.CLASS_NAME,"next")
    next_button.click()
    time.sleep(3)

quotes_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quotes_tags:
    quote=i.text
    quotes.append(quote)
authors_tags=driver.find_elements(By.CLASS_NAME,"author")
for i in authors_tags:
    author=i.text
    authors.append(author)
types_tags=driver.find_elements(By.CLASS_NAME,"tags")
for i in types_tags:
    typ=i.text
    types.append(typ)
    
df=pd.DataFrame({'Quotes':quotes,'Authors':authors,'Types of Quotes':types})
df.head(1000)
```

```python
# Q.9  a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/. 
# This task will be done in following steps: 
# 1. First get the webpagehttps://www.jagranjosh.com/ 
# 2. Then You have to click on the GK option 
# 3. Then click on the List of all Prime Ministers of India 
# 4. Then scrap the mentioned data and make theDataFrame. 

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.jagranjosh.com/")
gk=driver.find_element(By.XPATH,"/html/body/div/header/nav/div/div/div[3]/ul/li[3]/a")
gk.click()
time.sleep(5)
pm=driver.find_element(By.XPATH,"//a[@href='https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1']")
pm.click()

name=[]
born_dead=[]
term=[]
remarks=[]

name_tags=driver.find_elements(By.XPATH,'//div[@class="TableData"]/table/tbody/tr/td[2]/div')
for i in name_tags:
    n=i.text
    name.append(n)
born_tags=driver.find_elements(By.XPATH,'//div[@class="TableData"]/table/tbody/tr/td[3]/div')
for i in born_tags:
    born=i.text
    born_dead.append(born)
term_tags=driver.find_elements(By.XPATH,'//div[@class="TableData"]/table/tbody/tr/td[4]')
for i in term_tags:
    t=i.text
    term.append(t)
remark_tags=driver.find_elements(By.XPATH,'//div[@class="TableData"]/table/tbody/tr/td[5]/div')
for i in remark_tags:
    remark=i.text
    remarks.append(remark)

df=pd.DataFrame({'Name':name,'Born-Dead':born_dead,'Term of Office':term,'Remarks':remarks})
df
```

```python
# Q.10 Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/ 
# This task will be done in following steps: 
# 1. First get the webpage https://www.motor1.com/ 
# 2. Then You have to type in the search bar ’50 most expensive cars’ 
# 3. Then click on 50 most expensive cars in the world.. 
# 4. Then scrap the mentioned data and make the dataframe.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.motor1.com/")
time.sleep(5)
product=driver.find_element(By.XPATH,"/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input")
product.send_keys('50 most expensive cars')
search=driver.find_element(By.XPATH,'//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]')
search.click()
time.sleep(3)
cars=driver.find_element(By.XPATH,"/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]/h3/a")
cars.click()

name=[]
price=[]

name_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tags[0:50]:
    n=i.text
    name.append(n)
price_tags=driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
for i in price_tags[0:50]:
    p=i.text
    price.append(p)
    
df=pd.DataFrame({'Car Name':name,'Price':price})
df
```
