```python
#1 A python program to display all the header tags from wikipedia.org and make data frame.

from bs4 import BeautifulSoup
import requests
import pandas as pd
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup=BeautifulSoup(page.content)
headers=[]
for i in soup.find_all('h2',class_="mp-h2"):
    headers.append(i.text)
df=pd.DataFrame({'Header':headers})
df
```

```python
#2 A python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
# from https://presidentofindia.nic.in/former-presidents.htm and make data frame.
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
import regex as re
page=requests.get('https://presidentofindia.nic.in/former-presidents')
soup=BeautifulSoup(page.content)
presidents=[]
name=[]
term=[]
pattern=r"\n"
for i in soup.find_all('div',class_="desc-sec"):
    presidents.append(i.text)
    p=re.split(pattern,i.text)
    name.append(p[1])
    term.append(p[2])
df=pd.DataFrame({'Name':name,'Term of Office':term})
df
```

```python
#5 A python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
# make data frame
# i) Headline
# ii) Time
# iii) News Link
from bs4 import BeautifulSoup
import requests
import pandas as pd
page=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
headline=[]
time=[]
link=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
for j in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(j.text)
for k in soup.find_all('a',class_="LatestNews-headline"):
    link.append(k['href'])
df=pd.DataFrame({'HeadLine':headline,'Time':time,'News Link':link})
df
```

```python
#6 A python program to scrape the details of most downloaded articles from AI in last 90
# days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
# Scrape below mentioned details and make data frame
# i) Paper Title
# ii) Authors
# iii) Published Date
# iv) Paper URL
from bs4 import BeautifulSoup
import requests
import pandas as pd
page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup=BeautifulSoup(page.content)
title=[]
author=[]
date=[]
url=[]
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    title.append(i.text)
for j in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(j.text)
for k in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    date.append(k.text)
for l in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    url.append(l['href'])
df=pd.DataFrame({'Paper Title':title,'Authors':author,'Published Date':date,'Ppaer URL':url})
df
```

```python
#7 A python program to scrape mentioned details from dineout.co.in and make data frame
# i) Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image URL
from bs4 import BeautifulSoup
import requests
import pandas as pd
page=requests.get('https://www.dineout.co.in/bangalore-restaurants/welcome-back')
soup=BeautifulSoup(page.content)
name=[]
cuisine=[]
location=[]
rating=[]
url=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    name.append(i.text)
for j in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(j.text)
for k in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(k.text)
for l in soup.find_all('div',class_=["restnt-rating rating-4","restnt-rating rating-5"]):
    rating.append(l.text)
for m in soup.find_all('img',class_="no-img"):
    url.append(m['data-src'])
df=pd.DataFrame({'Restaurant Name':name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image URL':url})
df
```
